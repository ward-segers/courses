## Basic concepts in Data Science

- *variables*: general property of an object, allows to distinguish objects
- *values*: specific property, interpretation of that object
  - Example: variable: gender, value: male

### Measurement levels

> They refer to the types of variables in a dataset. And determine how data can be analyzed, visualized and interpreted.

- They help determine the most suitable method for analysis:
  - visualization methods (Measurement levels help determine the most appropriate way to visualize the data)
  - central tendency and dispersion (different levels of measurement allow for different statistical summaries)
  - examine the relationship between variables

Types of measurement levels:

- ***Qualitative***: 
  - not necessarily numeric values.
  - limited number of values
  - categories:
    - *nominal* categories: e.g. gender, race, country, shape
    - *ordinal* Order, rank: e.g. military rank, level of education
- ***Quantitative***:
  - Number with a unit of measurement
  - Many values, often unique
  - categories:
    - *interval*: no fixed zero point ==> no proportion. e.g. C°, F°
    - *ratio*: absolute zero point ==> proportions. e.g. distance (m), energy (j), weight (kg)

> *Quantitative* variables often contain the results of a measurement

### Relationship between variables

> Variables are related if there values change **systematically**

#### Causal relationships

- Frustration leads to aggression
- Alcohol leads to decreased alertness

> *Cause* is the independent variable
> *Consequence* is the dependent variable

## Sample testing

- *population*: the collection of all objects/people/... you want to investigate
- *sample*: a subset of the population from which measurements will be taken

> Under certain circumstances the results for the sample are representative for the population
> A sample is easier to analyze than an entire population

#### Sample types

- *Random sample*: every element from the population has an equal chance of being included in the sample
- *Non-random sample*: the elements for the sample are not randomly selected.
  - Objects that are collected easily are more likely to be selected (**convenience sampling**)
- *Stratified sampling*: population is devided into **strata** based on one or more *stratified variables*. A random sample is then taken from each stratum, either proportionally or equally, depending on the study's goals.
  - Examples: 
    - Gender: Male, Female
    - Age Groups: 18-25, 26-35, 36-50, 51+
    - Income levels: Low, Medium, High

### Possible errors

Measurement in a sample will typically deviate from the value in the entire population and is therefor prone to errors.

These errors can be:
- Accidental or systematic
- Sampling-error or non-sampling-error

#### Sampling-errors

- Accidental sampling errors are pure coincidence
- Systematic sampling errors
  - Online survey: people without internet are excluded
  - Street survey: only who is currently walking there
  - Voluntary survey: only interesting parties participate

#### Non-sampling errors

- Accidental non-sampling errors
  - Incorrectly ticked answers
- Systematic non-sampling errors
  - Poor or non-calibrated equipment
  - Value can be influence by the fact that you measure
  - Respondents lie
